```python
import numpy as np
import json 
import os
```

```python
# Folder where the files are stored
folderpath = "/Users/martines/Desktop/PhD/USMEFT-HL-LHC/Data"
```

```python
for filename in os.listdir(f"{folderpath}/NCDY/pseudo_data_Zprime"):
    prefix, _ = filename.split(".")
    with open(f"{folderpath}/NCDY/pseudo_data_Zprime/{filename}") as file_:
        data = json.load(file_)
    np.savetxt(f"{folderpath}/Concha/NCDY/pseudo_data_Zprime/{prefix}.txt",
           np.transpose([data["data"][:-1], data["SM"][:-1]])
           )
```
