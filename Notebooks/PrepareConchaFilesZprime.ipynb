```python
import numpy as np
import json 
import os
```

```python
# Folder where the files are stored
folderpath = "/Users/martines/Desktop/PhD/USMEFT-HL-LHC/Data"
```

```python
# Load USMEFT predictions
with open(f"{folderpath}/NCDY/USMEFT-NCDY-HLLHC.json") as file_:
    usmeft_ncdy = json.load(file_)
    
with open(f"{folderpath}/CCDY/USMEFT-CCDY-HLLHC.json") as file_:
    usmeft_ccdy = json.load(file_)
```

```python
# Saving each term of the simulations
for term, predictions in usmeft_ncdy.items():
    np.savetxt(
        f"{folderpath}/Concha/Zprime/USMEFT/NCDY/NCDY-{term}.txt",
        predictions[:10]
    )

for term, predictions in usmeft_ccdy.items():
    np.savetxt(
        f"{folderpath}/Concha/Zprime/USMEFT/CCDY/NCDY-{term}.txt",
        predictions[:8]
    )
```

```python
# Data CCDY
np.savetxt(f"{folderpath}/Concha/Zprime/pseudo_data/CCDY/CCDY_pseudo_data.txt",
           np.transpose([usmeft_ccdy["SM"][:8], usmeft_ccdy["SM"][:8]])
           )
```

```python
for filename in os.listdir(f"{folderpath}/NCDY/pseudo_data_Zprime"):
    prefix, _ = filename.split(".")
    with open(f"{folderpath}/NCDY/pseudo_data_Zprime/{filename}") as file_:
        data = json.load(file_)
    np.savetxt(f"{folderpath}/Concha/Zprime/pseudo_data/NCDY/{prefix}.txt",
           np.transpose([data["data"][:10], data["SM"][:10]])
           )
```
