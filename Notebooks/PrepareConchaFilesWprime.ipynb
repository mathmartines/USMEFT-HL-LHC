```python
import numpy as np
import json 
import os
```

```python
# Folder where the files are stored
folderpath = "/Users/martines/Desktop/PhD/USMEFT-HL-LHC/Data"
```

```python
# Load USMEFT predictions
with open(f"{folderpath}/NCDY/USMEFT-NCDY-HLLHC-parton-level.json") as file_:
    usmeft_ncdy = json.load(file_)
    
with open(f"{folderpath}/CCDY/USMEFT-CCDY-HLLHC.json") as file_:
    usmeft_ccdy = json.load(file_)
```

```python
# Saving each term of the simulations
for term, predictions in usmeft_ncdy.items():
    np.savetxt(
        f"{folderpath}/Concha/Wprime/USMEFT/NCDY/NCDY-{term}.txt",
        predictions[:10]
    )

for term, predictions in usmeft_ccdy.items():
    np.savetxt(
        f"{folderpath}/Concha/Wprime/USMEFT/CCDY/CCDY-{term}.txt",
        predictions[:8]
    )
```

```python
for filename in os.listdir(f"{folderpath}/NCDY/pseudo_data_Wprime"):
    prefix, _ = filename.split(".")
    with open(f"{folderpath}/NCDY/pseudo_data_Wprime/{filename}") as file_:
        data = json.load(file_)
    np.savetxt(f"{folderpath}/Concha/Wprime/pseudo_data/NCDY/{prefix}.txt",
           np.transpose([data["data"][:10], data["SM"][:10]])
           )
```

```python
for filename in os.listdir(f"{folderpath}/CCDY/pseudo_data_Wprime"):
    prefix, _ = filename.split(".")
    with open(f"{folderpath}/CCDY/pseudo_data_Wprime/{filename}") as file_:
        data = json.load(file_)
    np.savetxt(f"{folderpath}/Concha/Wprime/pseudo_data/CCDY/{prefix}.txt",
           np.transpose([data["data"][:8], data["SM"][:8]])
           )
```
