```python
import numpy as np
import json 
import os
```

```python
# Folder where the files are stored
folderpath = "/Users/martines/Desktop/PhD/USMEFT-HL-LHC/Data"
```

```python
# Load USMEFT predictions
with open(f"{folderpath}/NCDY/USMEFT-NCDY.json") as file_:
    usmeft_ncdy = json.load(file_)
```

```python
# Saving each term of the simulations
for term, predictions in usmeft_ncdy.items():
    np.savetxt(
        f"{folderpath}/Concha/NCDY/USMEFT/NCDY-{term}.txt",
        predictions[:24]
    )
```

```python
with open(f"{folderpath}/NCDY/pseudo_data_SM.json") as file_:
    data = json.load(file_)
    np.savetxt(f"{folderpath}/Concha/NCDY/NCDY-pseudo_data_SM.txt",
           np.transpose([data["data"][:24], data["SM"][:24]])
           )
```
