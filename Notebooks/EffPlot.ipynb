```python
import json 
import numpy as np
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib.lines import Line2D
from matplotlib.ticker import AutoMinorLocator
import os
from scipy.stats import chi2

os.environ['PATH'] = f"/Library/TeX/texbin:{os.environ['PATH']}"
rc('text', usetex=True)
plt.rcParams['text.usetex'] = True
rc('font', **{'family': 'serif'})
plt.rcParams.update({'font.size': 20})
```

```python
# Read the file
filename = "/Users/martines/Desktop/PhD/USMEFT-HL-LHC/Data/CCDY/eff_study/eff_study_8.json"

# Load file 
with open(filename) as file_:
    analysis = json.load(file_)
```

```python
parton_level, detector_level = [], []
# Fix vectors 
for parton, detector in zip(analysis["parton"], analysis["detector"]):
    if detector is not None and detector > 0 :
        parton_level.append(parton)
        detector_level.append(detector)
```

```python
# 2d hist plot 
plt.figure()

h = plt.hist2d(
    parton_level,
    detector_level,
    bins=20,
    range=[[100, 6200], [100, 6200]],
    cmap="Greys"
)

plt.colorbar(h[3])  # Add colormap on right

plt.xlim(100, 6200)
plt.ylim(100, 6200)

ax = plt.gca()  # Get current axes
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(axis='x', pad=17, labelsize=20, width=1)
ax.tick_params(axis='y', pad=17, labelsize=20, width=1)
ax.tick_params(axis="both", which="minor", top=True, right=True, length=3, direction="in", width=1)
ax.tick_params(axis="both", which="major", top=True, right=True, length=6, direction="in", labelsize=20, pad=10)
ax.annotate(r"$ m_{\ell\nu} >$ 4.0 TeV", xy=(15, 235), fontsize=20, xycoords="axes points")

plt.xlabel(r"$m_T^\mathrm{parton}$ [GeV]")
plt.ylabel(r"$m_T^\mathrm{recons}$ [GeV]")

plt.savefig(f"/Users/martines/Desktop/PhD/USMEFT-HL-LHC/Data/CCDY/eff_study/bin_8.pdf", format="pdf", bbox_inches="tight", dpi=300)
```
